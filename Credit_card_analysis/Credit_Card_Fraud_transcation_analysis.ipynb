```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
#load the dataset to pandas Dataframe
df = pd.read_csv('creditcard.csv')
df.head(10)
```

```python
df.info()
```

```python
#check missing values in the data set in each column
df.isnull().sum()
```

```python
#distribution of Legit and fraudulant transactions
df['Class'].value_counts()
```

# This dataset is highly unbalanced
0--> Normal Transaction
1--> Fraudulant transaction

```python
#separating the Normal and Fradulant transcation to two different variables
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the values of both transactions
df.groupby('Class').mean()
```

```python
#Undersampling

#Build a sample dataset containing similar distribution of Legit and Fraudulant dataset

#We know that Legit has 284315 data points and Fraud has only 492
#So we will take a random sample of 492 datapoint s from Legit and concatenate with Fraud datapoints
#So that we have equally distributed dataset with both Legit anf Fraud datapoints

legit_sample = legit.sample(n=492)
```

```python
#concatinate two variables legit_sample nd fraud
df_concat = pd.concat([legit_sample, fraud], axis=0)
```

```python
df_concat['Class'].value_counts()
```

```python
df_concat.groupby('Class').mean()
```

```python
df_concat.head()
```

```python
#splitting the dataset into features nad targets

X = df_concat.drop(columns='Class', axis=1)
Y = df_concat['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training and test data
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, stratify=Y ,random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
#1. Train our data set
#2. Check accuracy
#Logistic Regression
reg = LogisticRegression()
```

```python
#fit the training data to regression variable
reg.fit(x_train, y_train)
```

```python
#parse x_train data to predict library and check the accuracy of the prediction
x_train_prediction = reg.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy on Training data:', training_data_accuracy)
```

```python
#check the prediction and accuracy of test data
x_test_prediction = reg.predict(x_test)
testing_data_acccuracy = accuracy_score(x_test_prediction, y_test)
print('Accuracy on testing data: ', testing_data_acccuracy)
```
